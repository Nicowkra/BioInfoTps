```python
def protRandom(l):
    aminoacidos = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y']    
    seq = ""
    for i in range(l):
        seq += random.choice(aminoacidos)
    return seq
    
tamaño = input("Tamaño de la secuencia: ")
protRand = protRandom(int(tamaño))
print(protRand)
```

```python
import numpy as np

def protPonderada(l):
    frecuencias = {
        'A': 0.08, 'C': 0.01, 'D': 0.05, 'E': 0.06, 'F': 0.04, 
        'G': 0.07, 'H': 0.02, 'I': 0.06, 'K': 0.06, 'L': 0.10, 
        'M': 0.02, 'N': 0.04, 'P': 0.05, 'Q': 0.04, 'R': 0.05, 
        'S': 0.06, 'T': 0.05, 'V': 0.10, 'W': 0.01, 'Y': 0.03
    }
    
    aminoacidos = list(frecuencias.keys())
    prob = list(frecuencias.values())
    
    sumaProb = sum(prob)
    if not np.isclose(sumaProb, 1.0):
        raise ValueError(f"Las frecuencias suman {sumaProb}")
        
    seq = ""
    for i in range(l):
        seq += np.random.choice(aminoacidos, p=prob)
    return seq

protPond = protPonderada(1000)
print(protPond)
```

```python
from collections import Counter
import matplotlib.pyplot as plt
from Bio.Seq import Seq

def traducir(seq): #Funcion para traducir las secuencias de ADN a aminoacidos
    seq_obj = Seq(seq)
    if set(seq.upper()).issubset({'A', 'C', 'G', 'T'}):
        return str(seq_obj.translate(to_stop=True))
    return seq
def traducirRand(seq): #Si uso la funcion traducir, frena cuando encuentra un codon stop, pero al ser aleatorio puede tener uno en el medio
    seq_obj = Seq(seq)
    if set(seq.upper()).issubset({'A', 'C', 'G', 'T'}):
        return str(seq_obj.translate(to_stop=False)).replace("*", "")
def borrarFasta(seq): #Funcion para borrar el header de Fasta
    lineas = seq.strip().split('\n')
    secuenciaLimpia = "".join(lineas[1:]).replace(" ", "").replace("\r", "")
    return secuenciaLimpia
def calcFreq(secuencia):
    AA = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y']    
    conteo = Counter(secuencia)
    aminoacidos = sorted(conteo.keys())
    largoTotal = len(secuencia)
    
    freq = {}
    for aa in AA:
        if largoTotal > 0:
            freq[aa] = round((conteo.get(aa, 0) / largoTotal) * 100, 2)
        else:
            freq[aa] = 0.0
            
    return freq

freqDict = calcFreq(protPond)
print(freqDict)
```

```python
def graficarFreq(freqDict, titulo=""):
    plt.figure(figsize=(10, 5))
    aa = list(freqDict.keys())
    freqs = list(freqDict.values())
    plt.bar(aa, freqs, color='skyblue', edgecolor='navy')
    
    plt.xlabel('Aminoácidos')
    plt.ylabel('Frecuencia (%)')
    plt.title(titulo)
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.show()
```

```python
#b)
def extraerExoIntro(record): # Extraemos los intrones y exones usando la locacion de regiones codificantes
    listaExones = []
    listaIntrones = []

    for feature in record.features:
        if feature.type == 'CDS':
            partes = list(feature.location.parts)
            
            for parte in partes:
                inicioExon = int(parte.start)
                finExon = int(parte.end)
                listaExones.append([inicioExon, finExon])
            
            for i in range(len(partes) - 1):
                inicioIntron = int(partes[i].end)
                finIntron = int(partes[i+1].start)
                
                if inicioIntron < finIntron:
                    listaIntrones.append([inicioIntron, finIntron])
            
            break 
    return listaExones, listaIntrones

colageno = getGen("NG_007400.1")
lacz = getGen("V00296.1")
              
misExones, misIntrones = extraerExoIntro(colageno)

exones= []
intrones = []
for secu in misExones:
    exones.append(str(colageno.seq[secu[0]:secu[1]]))
for secu in misIntrones:
    intrones.append(str(colageno.seq[secu[0]:secu[1]]))
    
exonesTotal = "".join(exones)
intronesTotal = "".join(intrones)
seqLacz = str(lacz.seq)

protExon = traducir(exonesTotal)
protIntron = traducir(intronesTotal)
protLacz = traducir(seqLacz)

graficarFreq(calcFreq(protIntron), titulo="Intrón de Colágeno")
graficarFreq(calcFreq(protExon), titulo="Exón de Colágeno")
graficarFreq(calcFreq(protLacz), titulo="lacZ de E. coli")
```

```python
#a)
Entrez.email = email

listaS = []   
listaM = []   
listaL = []   
# separo por longitud de proteina
for i in range(0, len(idList), 500):
    bloqueActual = idList[i : i + 500]
    idConsultas = ",".join(bloqueActual)    
    handle = Entrez.esummary(db="protein", id=idConsultas)
    resumenes = Entrez.read(handle)
    handle.close()
    
    for r in resumenes:
        largo = int(r['Length'])
        idActual = r['Id']
                
        if largo <= 250:
            listaS.append(idActual)
        elif 250 < largo <= 500:
            listaM.append(idActual)
        else:
            listaL.append(idActual)
```

```python
import io
def descLista(lista): #Descarga todas las secuencias de cada lista 
    secuencias = []
    for i in range(0, len(lista), 200):
        bloque = lista[i : i + 200]
        idsConsulta = ",".join(bloque)
        handle = Entrez.efetch(db="protein", id=idsConsulta, rettype="fasta", retmode="text")
        for registro in SeqIO.parse(io.StringIO(handle.read()), "fasta"):
                secuencias.append(str(registro.seq))
            
        handle.close()
    return secuencias


fig = plt.figure(figsize=(15, 10))
gs = fig.add_gridspec(2, 4)

axes_custom = [
    fig.add_subplot(gs[0, 0:2]), # Fila 0, ocupa columnas 0 y 1 (Arriba Izquierda)
    fig.add_subplot(gs[0, 2:4]), # Fila 0, ocupa columnas 2 y 3 (Arriba Derecha)
    fig.add_subplot(gs[1, 1:3])  # Fila 1, ocupa columnas 1 y 2 (CENTRADO)
]

listas = [listaS, listaM, listaL]
medida = ["0-250 AA", "250-500 AA", "+500 AA"]

for i, lista in enumerate(listas):
    freqDict = calcFreq("".join(descLista(lista)))
    
    ax = axes_custom[i]
    aa = list(freqDict.keys())
    freqs = list(freqDict.values())
    
    ax.bar(aa, freqs, color='skyblue', edgecolor='navy')
    ax.set_xlabel('Aminoácidos')
    ax.set_ylabel('Frecuencia (%)')
    ax.set_title(f"Distribución de Secuencias de longitud entre {medida[i]}")
    ax.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

Decidimos dividir las proteínas en 3 subgrupos: las que tienen entre 0 y 250 AA, 250 y 500 AA y las que tienen más de 500 AA. Este rango fue arbitrario, tomado para que cada grupo tuviera una cantidad de proteínas similar. Al analizar estos grupos, graficando su distribución, no pudimos obtener diferencias significativas entre ellos. Sin embargo, sí notamos que estas distribuciones se asemejan a la aleatoria calculada antes. Esto no es algo nuevo, ya había ocurrido con la secuencia media que calculamos antes. Es por esto que nuestra conclusión es que los 3 grupos tienen una variedad y un tamaño lo suficientemente grande como para representar una distribución igual a la aleatoria.
