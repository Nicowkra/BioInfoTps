# Objetivo 4

### a) Partiendo de la secuencia de C. tetani, intente descubrir una proteína homóloga en humanos (idealmente la sGC) usando a) BLAST, b) PSI-BLAST, c) PFAM y d) PROSITE. Describa cómo procedió en cada caso, y si llegó al resultado deseado ¿Cuál es el método más sensible? ¿Cuál es el más rápido?

- BLAST: No se encontraron hits significativos correspondientes a proteínas humanas. Esto probablemente se debe a la gran distancia evolutiva entre bacterias y mamíferos, junto con las diferencias en la arquitectura global de dominios. BLAST resulta adecuado para detectar homólogos cercanos, pero pierde sensibilidad frente a relaciones evolutivas remotas.
- PSI-BLAST: Para detectar homólogos remotos, PSI-BLAST requiere construir iterativamente un perfil probabilístico a partir de secuencias relacionadas. Idealmente, esto debería realizarse contra toda la base de datos no redundante para incorporar secuencias puente evolutivas. Sin embargo, al intentar este procedimiento, el servidor de NCBI abortó la ejecución debido al límite de uso de CPU.  
Como estrategia alternativa, se aplicó un filtro taxonómico para Homo sapiens desde la primera iteración con el objetivo de reducir el costo computacional, pero bajo estas condiciones no se logró recuperar la sGC humana.
- PFAM: Se utilizó la entrada correspondiente al modelo oculto de Markov del dominio HNOB. Mediante el filtrado taxonómico para Homo sapiens, se identificaron aproximadamente 10 secuencias correspondientes a subunidades de la sGC. Este método permitió detectar exitosamente la relación evolutiva entre el dominio HNOB bacteriano y proteínas humanas.
- PROSITE: Se utilizaron los patrones compuestos obtenidos en el ítem 3e, aplicando además el filtro taxonómico de Homo sapiens. Mediante esta estrategia se recuperaron secuencias correspondientes a la sGC humana.

Los métodos más sensibles para la detección de homólogos remotos son Pfam y PSI-BLAST (teóricamente, si los recursos de cómputo permiten las iteraciones). Esto se debe a que abandonan las matrices de sustitución estáticas como las que usa BLAST y emplean modelos probabilísticos como Modelos Ocultos de Markov y PSSMs. Esto logra encontrar las similitudes estructurales, aunque la identidad sea baja.    
Los métodos más rápidos fueron PROSITE y PFAM. En estas bases de datos secundarias, gran parte del procesamiento computacional ya se encuentra precalculado, por lo que el análisis consiste principalmente en comparar la secuencia problema contra signatures previamente construidas e indexadas, reduciendo considerablemente el costo computacional.
