## 2.2)

### Para analizar el resultado, complete el script (o código de python) a partir del que mostramos más abajo, para generar una tabla que muestre las siguientes columnas: query, hit, cobertura de la query, cobertura del hit, identidad y de qué especie es la proteína hit (donde “busqueda.blast” debe ser el resultado de la búsqueda anterior)

```python
import Bio.SearchIO as bpio
import pandas as pd

resultadosList = list(bpio.parse('Archivos/busqueda.blast', 'blast-xml'))
datosBlast = []

for queryResult in resultadosList:
    queryLen = queryResult.seq_len
    
    for hit in queryResult.hits:
        hitLen = hit.seq_len
        
        especieNombre = "Desconocida"
        if '[' in hit.description and ']' in hit.description:
            especieNombre = hit.description.split('[')[-1].split(']')[0]
            
        # Nos quedamos con el mejor alineamiento de cada hit
        if len(hit.hsps) > 0:
            hspActual = hit.hsps[0]
            
            covQuery = (hspActual.query_span / queryLen) * 100
            covHit = (hspActual.hit_span / hitLen) * 100
            identidadPorcentaje = (hspActual.ident_num / hspActual.aln_span) * 100
            
            datosBlast.append({
                "Query": queryResult.id,
                "Hit": hit.id,
                "Cov_Query(%)": round(covQuery, 2),
                "Cov_Hit(%)": round(covHit, 2),
                "Identidad(%)": round(identidadPorcentaje, 2),
                "Especie": especieNombre
            })

dfResultados = pd.DataFrame(datosBlast)
dfFiltrado = dfResultados[(dfResultados['Cov_Hit(%)'] >= 70.0)]
dfOrdenado = dfFiltrado.sort_values(by='Identidad(%)', ascending=False).reset_index(drop=True)

dfOrdenado
```

# 2.3)

### a) Realice un blast del proteoma de N315 contra la base de enterotoxinas y desarrolle un script que levantando esa información indique cuáles son las enterotoxinas que posee (o sea están presentes en) el proteoma. (Posiblemente obtenga varios hits para cada proteína)

```python
import Bio.SearchIO as bpio
from Bio import BiopythonParserWarning
import pandas as pd
import warnings
warnings.simplefilter('ignore', BiopythonParserWarning)

resultadosList = list(bpio.parse('Archivos/enterotoxinas.blast', 'blast-xml'))
datosBlast = []

for queryResult in resultadosList:
    queryLen = queryResult.seq_len
    for hit in queryResult.hits:
        hitLen = hit.seq_len
        
        if len(hit.hsps) > 0:
            hspActual = hit.hsps[0]
            
            covQuery = (hspActual.query_span / queryLen) * 100
            covHit = (hspActual.hit_span / hitLen) * 100
            identidadPorcentaje = (hspActual.ident_num / hspActual.aln_span) * 100
            
            datosBlast.append({
                "Proteina_N315_Query": queryResult.id,
                "Enterotoxina_Hit": hit.id,
                "Descripción_Hit": hit.description, 
                "Cov_Query(%)": round(covQuery, 2),
                "Cov_Hit(%)": round(covHit, 2),
                "Identidad(%)": round(identidadPorcentaje, 2)
            })

dfEnterotoxinas = pd.DataFrame(datosBlast)

dfFiltrado = dfEnterotoxinas[dfEnterotoxinas['Identidad(%)'] >= 30.0].sort_values(by='Identidad(%)', ascending=False).reset_index(drop=True)
toxinasPresentes = dfFiltrado['Enterotoxina_Hit'].unique()

print(f"Se encontraron matches contra {len(toxinasPresentes)} secuencias representativas de enterotoxinas.")
print("Las enterotoxinas presentes son:")
for toxina in toxinasPresentes:
    print(f"- {toxina}")

dfFiltrado
```

### b) Programe algún criterio de decisión en función de la identidad/cobertura. En caso de que más de un hit cumpla las expectativas, indique que encontró más de una proteína

```python
# 2. Definimos nuestro Criterio de Decisión:
umbralIdentidad = 80.0
umbralCobertura = 80.0

dfEstricto = dfEnterotoxinas[
    (dfEnterotoxinas['Identidad(%)'] >= umbralIdentidad) &
    (dfEnterotoxinas['Cov_Hit(%)'] >= umbralCobertura)
]

resultadosFinales = []
gruposQuery = dfEstricto.groupby('Proteina_N315_Query')

for queryId, dfGrupo in gruposQuery:
    dfGrupoOrdenado = dfGrupo.sort_values(by='Identidad(%)', ascending=False)
    if len(dfGrupoOrdenado) == 1:
        mejorHit = dfGrupoOrdenado.iloc[0]['Enterotoxina_Hit']
        resultadosFinales.append({
            "Proteina_N315": queryId,
            "Enterotoxina_Asignada": mejorHit,
            "Criterio_Decision": "Hit unico válido"
        })
    else:
        # Si mas de un hit cumple las expectativas
        nombresToxinas = " | ".join(dfGrupoOrdenado['Enterotoxina_Hit'].tolist())
        resultadosFinales.append({
            "Proteina_N315": queryId,
            "Enterotoxina_Asignada": nombresToxinas,
            "Criterio_Decision": "Se encontró más de una proteína"
        })

dfDecision = pd.DataFrame(resultadosFinales)
dfDecision
```

Dado el alto grado de homología dentro de la familia de las enterotoxinas, se observó que una única secuencia query alineaba con múltiples variantes. Para decidir, se programó un filtro estableciendo un umbral de $\ge$ 80% de Identidad y $\ge$ 80% de Cobertura.  
En aquellos casos donde múltiples toxinas superaron el umbral para una misma query, el algoritmo lo clasificó indicando que se encontró más de una proteína válida.
